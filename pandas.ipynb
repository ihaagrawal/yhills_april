```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.hist(x=df["Age"])
```

```python
plt.hist(x=df["Age"],color="red")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(x=df["Fare"],color="black")
plt.title("Distribution of Fare",fontsize=20,color="blue")
plt.xlabel("Fare",fontsize=10,color="blue")
plt.ylabel("Frequency",fontsize=10,color="blue")
plt.show()
```

```python
sns.boxplot(df["Age"],color="black")
plt.title("Distribution of Age",color="blue", fontsize=20)
plt.xlabel("Age")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Distribution of Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="barh")
plt.title("Distribution of Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Pclass"].value_counts().plot(kind="barh")
plt.title("Distribution of Pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```python
from matplotlib import cm
emb=df["Embarked"].value_counts()
keys=emb.keys().to_list()
counts=emb.to_list()
cs=cm.Set1([3,5,7,9])
plt.pie(x=counts,labels=keys,autopct="%1.1f%%",colors=cs)
plt.show()
```
